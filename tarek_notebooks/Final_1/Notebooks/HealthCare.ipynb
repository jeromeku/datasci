# Import Necessary Packages and Modules

```python
import pandas as pd # for reading csv files
import pylab as pl # for plotting
from IPython.display import Image # for viewing images in ipython notebook
import numpy as np
```

# Load Data

Read csv file

```python
csv_file = '../data/healthcare2.csv'
hc = pd.read_csv(csv_file)
```



Select some columns for plotting

```python
X = hc[['AGE','INCOME']]
X = np.array(X)
Y = np.array(hc['Y'])
```

# Plot Data

```python
# When Y is 1, then the TurboTax Filer has HealthCare Insurance
idx_positive = Y==1
plt = pl.scatter(X[idx_positive, 1], X[idx_positive, 0], c='b', marker='s', alpha=0.5)
# When Y is 0, then the TurboTax Filer does not have HealthCare Insurance
idx_negative = Y==0
plt = pl.scatter(X[idx_negative, 1], X[idx_negative, 0], c='r', marker='o', alpha=0.5)

#add Bob
# age = 25
# income = 10000
# plt = pl.scatter(income, age, c='k', marker='v', s=300)
# plt = pl.text(income-1500, age+1, 'Bob', fontsize = 15)

#make figure look nice
plt.axes.set_xlim([0,100000])
plt.axes.set_ylim([17,40])

pl.xlabel('Income')
pl.ylabel('age')
pl.savefig('toremove.png', dpi= 200)
pl.close()
Image('toremove.png')
```
