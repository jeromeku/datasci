# Import Necessary Packages and Modules

```python
import pandas as pd
import pylab as pl
from IPython.display import Image
import numpy as np


from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression

# import libraries for spliting
import random
```

# Load Data

Read csv file

```python
csv_file = '../data/healthcare2.csv'
hc = pd.read_csv(csv_file)
#hc
```

```python
X = hc[['AGE','INCOME']]
X = np.array(X)
Y = np.array(hc['Y'])
```

# Split Data into Training and Testing Sets

```python
# shuffle data
n_samples, n_features = X.shape
p = range(n_samples)
random.seed(0)
random.shuffle(p)
X, Y = X[p], Y[p]

half = int(n_samples/2)

X_train, y_train = X[:half], Y[:half]
X_test, y_test = X[half:], Y[half:]
```

# Train Classifier on Training Data Set

```python

# ==================================
# PREPARE THE CLASSIFIER
# ==================================

# Define Classifier Object
clf = LinearSVC(penalty='l1', C=0.001, dual=False)


# ==================================
# FIT THE MODEL
# ==================================
# Fit the Data using the Classifier at hand
clf.fit(X_train, y_train)
```

# Plot Testing Data

```python


# Color Decision Areas for each class

h = 1  # step size in the mesh
# create a mesh to plot in
age_min, age_max = X_test[:, 0].min() - 1, X_test[:, 0].max() + 1
income_min, income_max = X_test[:, 1].min() - 1, X_test[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(age_min, age_max, 0.1),
                     np.arange(income_min, income_max, 300))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt = pl.contourf(yy, xx, Z, cmap=pl.cm.Paired, alpha=0.8)


# Plot Boundary line if it exists
try:
    x_min, x_max = X_test[:, 0].min(), X_test[:, 0].max()
    xx = np.linspace(x_min, x_max)
    w = clf.coef_.ravel()
    a = -w[0] / w[1]
    yy = a * xx - clf.intercept_ / w[1]
    h0 = pl.plot(yy, xx, 'k')
except Exception as e:
    pass 

#=======================
# Plot the data
#=======================

# When Y is 1, then the TurboTax Filer has HealthCare Insurance
idx_positive = y_test==1
plt = pl.scatter(X_test[idx_positive, 1], X_test[idx_positive, 0], c='b', marker='s', alpha=0.5)
# When Y is 0, then the TurboTax Filer does not have HealthCare Insurance
idx_negative = y_test==0
plt = pl.scatter(X_test[idx_negative, 1], X_test[idx_negative, 0], c='r', marker='o', alpha=0.5)

plt = pl.scatter(10000, 25, c='k', marker = 'v', s=300)

plt.axes.set_xlim([0,100000])
plt.axes.set_ylim([17,40])
pl.xlabel('Income')
pl.ylabel('age')
pl.savefig('toremove.png', dpi= 200)
pl.close()
Image('toremove.png')
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
yclf = clf.predict(X_test)
print('Report:\n')
target_names = ['noHelathCare', 'hasHealthCare']
print(classification_report(y_test,yclf, target_names=target_names))
```
